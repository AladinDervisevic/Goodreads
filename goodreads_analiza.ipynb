```python
import pandas as pd

pd.options.display.max_rows = 20
```

```python
books = pd.read_csv("goodreads-processed-data\\goodreads.csv")
```

```python
books[['title', 'average_rating', 'score']].sort_values(['average_rating', 'score'], ascending = [False, False])
```

So the best book by average rating and score is Otuna's flute.

```python
books[['author', 'title']].groupby('author').count().sort_values('title', ascending = False).head(20)
```

The author that wrote the most books on this site is Stephen King.

```python
books[books.average_rating > 4.0].groupby('author').count().sort_values('title', ascending = False)
```

This is a spreadsheet of authors that wrote the most books which have an average rating above 4.0 .

```python
books[books.author == "Ernest Hemingway"].sort_values('average_rating', ascending = False)
```

So the best rated book by Hemingway is "The Short Stories".

```python
books[books.author == "J.R.R. Tolkien"].sort_values('average_rating', ascending = False)
```

```python
books["metascore"] = books.ratings / books.average_rating
books.sort_values('metascore', ascending=False).head(10)
```

Based off of the new rating/grading system, these are the top rated books out there, which is not that surprising because Harry Potter is quite beloved by most.
